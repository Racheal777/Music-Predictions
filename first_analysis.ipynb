```python
#importing packages
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

#read the data with panda
data = pd.read_csv('music.csv')

#assign all columns except genre to the X variable
X = data.drop(columns=['genre'])

#assign the genre column to the y
y = data['genre']

#assign 20% of the data for testing
#first 2 variables are input for training and testing
#last 2 variables are output for training and testing
#80% for training the data, 20% for testing the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

#assign the decision  class to the model to be able to use for prediction
model = DecisionTreeClassifier()

#model.fit, grab the values of the X so it doesn’t give you a warning
# model.fit(X.values, y)
model.fit(X_train, y_train)

print(f" X test data are {X_test}")
print(f" y test data are {y_test}")

#make predictions with a few sample data of a male in his 30 and a woman in her 40 and the kind of music they will listen to
# predictions = model.predict([ [30, 1], [40, 0] ])
predictions = model.predict(X_test)

#predictions based on the data provided
predictions

score = accuracy_score(y_test, predictions)
score
```

```python
#importing packages
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

#read the data with panda
data = pd.read_csv('music.csv')

#assign all columns except genre to the X variable
X = data.drop(columns=['genre'])

#assign the genre column to the y
y = data['genre']

#assign the decision  class to the model to be able to use for prediction
model = DecisionTreeClassifier()

#model.fit, grab the values of the X so it doesn’t give you a warning
model.fit(X.values, y)

tree.export_graphviz(model, out_file = 'music_recommender.dot',
                    feature_names= ['age', 'gender'],
                    class_names =  sorted(y.unique()),
                    label = 'all',
                     rounded = True,
                     filled = True)
                     
```
